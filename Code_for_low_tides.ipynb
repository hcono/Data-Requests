## Jupyter notebook of code to get Lowest Tide Times
- works by getting the troughs from the water_level column
- output is a dataframe

```python
import pandas as pd
# use scipy find_peaks in the negative to find the troughs
from scipy.signal import find_peaks
```

```python
# get the file html, sourced in ERDDAP
file = 'https://erddap.marine.ie/erddap/tabledap/IMI-TidePrediction.csvp?time%2Clongitude%2Clatitude%2CstationID%2CWater_Level%2CWater_Level_ODM&time%3E=2023-01-01&time%3C=2024-01-01'

df = pd.read_csv(file)
```

```python
# use the negative to find the troughs
# this returns a tuple
troughs = find_peaks(-(df['Water_Level (metres)']))
```

```python
# the first index of the tuple has a list of indices
troughlist = troughs[0]
```

```python
# use take() to get the rows from the list of indices above
troughcheckdf = df.take(troughlist)
troughcheckdf
```
